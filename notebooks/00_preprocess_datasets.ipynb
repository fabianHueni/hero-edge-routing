# Dataset Preprocessing
This notebook preprocesses several datasets by adding a unique identifier column to each dataset and saving the modified datasets back to CSV files. Furthermore, it limits the number of items in the dataset to 1000 for quicker experimentation.

```python
import pandas as pd
```

# Prepare the spam/ham email dataset

```python
dataset = pd.read_csv('../data/spam_ham_dataset.csv')
```

```python
# add column with row id
dataset['ID'] = range(1, len(dataset) + 1)
```

```python
# add the column to the first position
cols = dataset.columns.tolist()
cols = cols[-1:] + cols[:-1]
dataset = dataset[cols]
```

```python
dataset.head()
```

```python
# filter out elements where the text has more than 10000 characters (otherwise webgpu is not able to process them)
dataset = dataset[dataset['Text'].str.len() <= 10000]
```

```python
# split the dataset in spam and ham and then sample from each class to have balanced dataset
spam = dataset[dataset['Spam'] == 1].sample(n=500, random_state=42)
ham = dataset[dataset['Spam'] == 0].sample(n=500, random_state=42)
dataset = pd.concat([spam, ham]).reset_index(drop=True)
```

```python
# random sorting of dataset
dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# overwrite ID column with new range after random sampling
dataset['ID'] = range(1, len(dataset) + 1)
```

```python
dataset.head(10)
```

```python
# store csv in dataset folder
dataset.to_csv('../data/spam_ham_dataset.csv', index=False)
```

# Prepare the AG news dataset

```python
dataset = pd.read_csv('../data/ag_news_test.csv')
```

```python
# add column with row id
dataset['ID'] = range(1, len(dataset) + 1)
```

```python
# add the column to the first position
cols = dataset.columns.tolist()
cols = cols[-1:] + cols[:-1]
dataset = dataset[cols]
```

```python
dataset.head()
```

```python
# sample from each class to have balanced dataset
class_world = dataset[dataset['Class Index'] == 1].sample(n=250, random_state=42)
class_sports = dataset[dataset['Class Index'] == 2].sample(n=250, random_state=42)
class_business = dataset[dataset['Class Index'] == 3].sample(n=250, random_state=42)
class_science = dataset[dataset['Class Index'] == 4].sample(n=250, random_state=42)
dataset = pd.concat([class_world, class_sports, class_business, class_science]).reset_index(drop=True)
```

```python
# random sorting of dataset
dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# overwrite ID column with new range after random sampling
dataset['ID'] = range(1, len(dataset) + 1)
```

```python
dataset.head(10)
```

```python
# store csv in dataset folder
dataset.to_csv('../data/ag_news_test.csv', index=False)
```

## Prepare the BoolQ dataset

```python
dataset = pd.read_csv('../data/boolq_validation.csv')
```

```python
dataset.head()
```

```python
# sample from each class to have balanced dataset
true_samples = dataset[dataset['answer'] == True].sample(n=500, random_state=42)
false_samples = dataset[dataset['answer'] == False].sample(n=500, random_state=42)
dataset = pd.concat([true_samples, false_samples]).reset_index(drop=True)
```

```python
# random sorting of dataset
dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# overwrite ID column with new range after random sampling
dataset['ID'] = range(1, len(dataset) + 1)
```

```python
dataset.head(10)
```

```python
# store csv in dataset folder
dataset.to_csv('../data/boolq_validation.csv', index=False)
```
